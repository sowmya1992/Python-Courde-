```python
### corelation is a statistical measure to quantify the degree of relationship between two or more variables. 
import pandas as pd

data = pd.read_csv("C:/Users/Sowmya/Desktop/Statistics/datasets/data_visualization.csv")## reading the data 
data.head(10)


```

```python
data.columns ### finding the column names 
```

```python
data.isnull().values.any() ## finding missing values 
```

```python
data.describe() ## finding the values for each column like min,max,std etc...
 
```

```python
corr = data.corr()
corr

```

```python
print(data[['length','height']].corr()) ## finding the relationship between length and height 

print(data[['interior','arrow']].corr()) ## interior and arrow 

print(data[['rho','twi']].corr()) ### Rho and twi 
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)
cax=ax.matshow(corr,cmap = 'coolwarm', vmin =-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation= 90)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
plt.show() 

### Negavtive corelation is shown in the graph below.
```

```python

import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)
cax=ax.matshow(data[['rho','twi']].corr(),cmap = 'coolwarm', vmin =-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation= 90)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
plt.show()   
```

```python
##Interpretation 

### We can see tat there is negative corelation between the variables.
##As Length increases height decreases and vice versa.
## As interior decreases arrow decreases and vice versa.
## AS rho increases twi decreases and vice versa.

```
